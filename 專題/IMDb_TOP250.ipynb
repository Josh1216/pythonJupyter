```python
#TOP250名稱,年分,時長,分級
import requests
import openpyxl
from bs4 import BeautifulSoup
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service

# 先更改chrome裡的設定
option = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values":{"notifications":2}}
option.add_experimental_option("prefs",prefs)

# 再套用在原本開啟的設定並覆蓋
service = Service(executable_path = "D:\\JOSHWANG\\.python.jupyter\\chromedriver.exe")
web = webdriver.Chrome(service=service,options=option)
web.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250")

wb=openpyxl.Workbook()
ws=wb.active
a1=[]
b1=[]
c1=[]
d1=[]
e1=[]
soup = BeautifulSoup(web.page_source,"html.parser")
for i in soup.find_all('div', class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-14dd939d-7 fjdYTb cli-title"):
    a1 = str(i.find('h3', class_="ipc-title__text").text)    
    metadata = i.find_next('div', class_="sc-14dd939d-5 cPiUKY cli-title-metadata")
    try:
        b1 = str(metadata.select('span')[0].text)
        c1 = str(metadata.select('span')[1].text)
        d1 = str(metadata.select('span')[2].text)
    except:
        b1 = str(metadata.select('span')[0].text)
        c1 = str(metadata.select('span')[1].text)
        d1 = "R-18"
        
    # 抓取評分的元素
    rating = i.find_next('span', class_="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating")
    e1 = str(rating.text)
    
    ws.append([a1 , b1 , c1 , d1 , e1]) 
    
wb.save("IMDb_TOP250.xlsx")
```
