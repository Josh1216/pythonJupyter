```python
import random #GPT

def flip_coin():
    # 產生隨機數，0代表反面，1代表正面
    return random.choice([0, 1])

def play_game(guess):
    # 玩家猜測的結果（0代表反面，1代表正面）
    if guess not in [0, 1]:
        return "請輸入有效的猜測：0代表反面，1代表正面"
    
    # 產生硬幣的結果
    coin_result = flip_coin()

    # 檢查猜測是否正確
    if guess == coin_result:
        return "恭喜，猜對了！硬幣是%s面。" % ("正" if coin_result == 1 else "反")
    else:
        return "很遺憾，猜錯了！硬幣是%s面。" % ("正" if coin_result == 1 else "反")

if __name__ == "__main__":
    print("歡迎來到猜硬幣遊戲！")
    while True:
        print("請輸入你的猜測：0代表反面，1代表正面，或輸入'exit'結束遊戲。")
        user_guess = input()
        if user_guess.lower() == 'exit':
            print("感謝遊玩，再見！")
            break

        try:
            user_guess = int(user_guess)
            result = play_game(user_guess)
            print(result)
        except ValueError:
            print("請輸入有效的猜測：0代表反面，1代表正面")
```
