```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import openpyxl

res=requests.get('https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword=%E5%A4%A7%E6%95%B8%E6%93%9A&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&order=12&asc=0&page=1&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1')
soup=BeautifulSoup(res.text)
wb=openpyxl.Workbook()
ws=wb.active

ws['A1']="職缺名稱"
ws['B1']="職缺連結"
ws['C1']="公司名稱"
ws['D1']="工作地區"
ws['E1']="薪資待遇"
ws['F1']="給薪方式"
ws['G1']="薪資下限"
ws['H1']="薪資上限"

page=1
while soup.find_all('article',class_="b-block--top-bord job-list-item b-clearfix js-job-item")!=[]:
    print('=======================')
    print('現在正在讀取第',page,'頁')
    #單頁爬蟲
    for job in soup.find_all('article',class_="b-block--top-bord job-list-item b-clearfix js-job-item"):
        a1 = str(job.a.text)
        b1 = str('https:'+job.a['href'])
        c1 = str(job.ul.a.text.strip())
        d1 = str(job.select('ul')[1].li.text)
        if  job.find('div',class_="job-list-tag b-content").select('span')!=[] and job.find('div',class_="job-list-tag b-content").select('span')[0].text=='待遇面議':
            e1 = str(job.find('div',class_="job-list-tag b-content").span.text)
            f1 = ""
            low_money = ""
            high_money = ""
        else:
            e1 = str(job.find('div',class_="job-list-tag b-content").a.text)
            
            f1 = e1[:2]
            strion = ''

            for word1 in e1:
                if word1 =='~' or word1.isdigit():
                    strion += word1
            if '~' in strion:        
                low_money = strion[:strion.find('~')]    
                high_money = strion[strion.find('~')+1:]
            else:
                low_money = strion
                high_money = strion

            low_money = int(low_money)
            high_money = int(high_money)
            
        ws.append([a1 , b1 , c1 , d1 , e1 , f1 , low_money , high_money])
    page+=1    
    res=requests.get('https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword=%E5%A4%A7%E6%95%B8%E6%93%9A&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&order=12&asc=0&page='+str(page)+'&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1')
    soup=BeautifulSoup(res.text)
    sleep(2)
    wb.save("excel測試20230807.xlsx")
```
